```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mydata_GP_years = mydata.groupby('Year')['GP'].mean()
```

## Stats development over the years ##

```python
mydata_GAA_years = mydata.groupby('Year')['GAA'].mean()
```

```python
plt.figure(figsize=(10,5))
plt.plot(mydata_GAA_years.index, mydata_GAA_years.values, marker='o')
plt.title('Average Goals Against Average (GAA) Over the Years')
plt.xlabel('Year')
plt.ylabel('Average GAA')
plt.show()
```

This data represents the goals scored on all goalkeepers by year, we can see a trend where before the 2000s there were relatively more goals scored against goalkeepers, but after the 2000s we can see that an improvement was made and fewer goals were being scored.
But don't get tricked! As we've seen earlier there is an underlying tendency that a lot of goalkeepers were signed in the 2000s but didn't get to play any games, therefore their stats are zeros which affect the averages of GAA lowering them.

```python
mydataPlayed_GAA_years = mydataPlayed.groupby('Year')['GAA'].mean()
```

```python
plt.figure(figsize=(10,5))
plt.plot(mydataPlayed_GAA_years.index, mydataPlayed_GAA_years.values, marker='o')
plt.title('Average Goals Against Average (GAA) Over the Years')
plt.xlabel('Year')
plt.ylabel('Average GAA')
plt.show()
```

Here's what the actual trend may look like. 
We can see before the 2000s they averaged almost 2 goals per game. Later, there was a lowest point in 2010 We're the average barely passes the mark of 1 goal per game. Then it improved to get above the 1.5 goals per game mark.
