```python
penalty_event_counts = mypenalties['Time'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
penalty_event_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Penalty Events Over Game Time')
plt.xlabel('Time in Minutes')
plt.ylabel('Number of Penalty Events')
plt.xticks(rotation=90)
plt.show()
```

```python
def convert_time_to_int(time_str):
    if '+' in time_str:
        base, extra = time_str.split('+')
        minutes = int(base.strip("'")) + 0.5
    else:
        minutes = int(time_str.strip("'"))
    return minutes
```

```python
mypenalties['Minute'] = mypenalties['Time'].apply(convert_time_to_int)
mypenalties
```

```python
intervals = [
    (1, 45),
    (45, 46),
    (46, 90),
    (90, 91)
]

labels = [
    '1-45', '45-46', '46-90', '90-91'
]
```

```python
mypenalties['Time_category'] = pd.cut(mypenalties['Minute'], bins=[start for start, end in intervals] + [intervals[-1][1]], labels=labels, right=False)
mypenalties
```

```python
time_category_counts = mypenalties['Time_category'].value_counts().sort_index()

plt.figure(figsize=(12, 8))
time_category_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Penalty Events Across Time Categories')
plt.xlabel('Time Category')
plt.ylabel('Number of Penalty Events')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
success_rate_per_interval = mypenalties.groupby('Time_category').apply(
    lambda x: (x['Outcome'] == 'Goal').mean() * 100
).reset_index(name='Success Rate')

success_rate_per_interval
```

```python
plt.figure(figsize=(12, 8))
plt.bar(success_rate_per_interval['Time_category'], success_rate_per_interval['Success Rate'], color='skyblue', edgecolor='black')
plt.title('Success Rate of Penalty Events Across Time Categories')
plt.xlabel('Time Category')
plt.ylabel('Success Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Player Performance ##

```python
mypenaltiesplayers = mypenalties[['Player', 'year']].drop_duplicates().reset_index(drop=True)
mypenaltiesplayers
```

```python
myplayers = pd.read_csv('all_players.csv')
```

```python
myplayers = myplayers[myplayers['Season'] == 'reg']
```

```python
#Standarized offensive and defensive midfielder positions
myplayers['POS'] = myplayers['POS'].replace(['M-D', 'D-M'], 'MD')
myplayers['POS'] = myplayers['POS'].replace(['M-F', 'F-M'], 'MO')
```

```python
# Set 'SOG%' to 0.0 where both 'SHTS' and 'SOG' are zero
myplayers.loc[(myplayers['SHTS'] == 0) & (myplayers['SOG'] == 0), 'SOG%'] = 0.0

final_missing_values_confirmation = myplayers.isnull().sum()
final_missing_values_confirmation
```

```python
#Select columns of interest
columnlist = ['Year', 'Player', 'POS', 'GP', 'PKG/A', 'SHTS', 'SOG', 'SC%', 'GWG', 'G', 'A', 'HmG', 'RdG',]

mysortedplayers = myplayers[columnlist]
mysortedplayers
```

```python
myplayerstats = pd.merge(mypenaltiesplayers, mysortedplayers, left_on=['Player', 'year'], right_on=['Player', 'Year'], how='inner')
myplayerstats
```

```python
myplayerstats.to_csv('players.csv', index=False)
```

```python
player_performance = mypenalties.groupby('Player').agg(
    penalties_taken=('Outcome', 'size'),
    penalties_scored=('Outcome', lambda x: (x == 'Goal').sum())
).reset_index()

player_performance['success_rate'] = (player_performance['penalties_scored'] / player_performance['penalties_taken']) * 100
player_performance_overall = player_performance.sort_values(by='penalties_taken', ascending=False)

player_performance_overall.head(30)
```

```python
penalty_data = mypenalties[['year', 'Player', 'Outcome']]

player_performance_year = penalty_data.groupby(['year', 'Player']).agg(
    penalties_taken=('Outcome', 'size'),
    penalties_scored=('Outcome', lambda x: (x == 'Goal').sum())
).reset_index()

player_performance_year['scoring_percentage'] = (player_performance_year['penalties_scored'] / player_performance_year['penalties_taken']) * 100


player_performance_year
```

```python
merged_players = pd.merge(myplayerstats, player_performance_year, how='inner', left_on=['year', 'Player'], right_on=['year', 'Player'])
merged_players
```

```python
position_analysis = merged_players.groupby('POS').agg(
    total_penalties_taken=('penalties_taken', 'sum'),
    total_penalties_scored=('penalties_scored', 'sum'),
    total_goals=('G', 'sum'),
    total_assists=('A', 'sum')
).reset_index()

position_analysis
```

```python
position_analysis.plot(kind='bar', x='POS', y=['total_penalties_taken', 'total_penalties_scored'])
plt.ylabel('Count')
plt.title('Position-wise Performance')
plt.show()
```

```python
position_analysis.plot(kind='bar', x='POS', y=['total_goals', 'total_assists'])
plt.ylabel('Count')
plt.title('Position-wise Performance')
plt.show()
```

```python
# Extract relevant columns and drop duplicates
player_years = merged_players[['year', 'Player']].drop_duplicates()

# Calculate the number of years each player has been taking penalties
player_years['penalty_years'] = player_years.groupby('Player')['year'].rank(method='dense').astype(int)

# Merge this information back with the original penalty_stats DataFrame
penalty_experience = pd.merge(merged_players, player_years, how='left', on=['year', 'Player'])

# Display the first few rows of the resulting DataFrame
penalty_experience.head(30)
```

```python
# Group by player and year to get the stats for each year
yearly_stats = penalty_experience.groupby(['penalty_years', 'Player']).agg(
    penalties_taken=('penalties_taken', 'sum'),
    penalties_scored=('penalties_scored', 'sum'),
    scoring_percentage=('scoring_percentage', 'mean')
).reset_index()

# Display the resulting DataFrame
yearly_stats
```

```python
# Analyze the performance based on experience
experience_analysis = yearly_stats.groupby('penalty_years').agg(
    avg_penalties_taken=('penalties_taken', 'mean'),
    avg_penalties_scored=('penalties_scored', 'mean'),
    avg_scoring_percentage=('scoring_percentage', 'mean')
).reset_index()

experience_analysis
```

```python
experience_analysis.plot(x='penalty_years', y='avg_scoring_percentage', kind='line')
plt.xlabel('Years of Experience')
plt.ylabel('Average Values')
plt.title('Performance Based on Experience')
plt.show()
```

```python
players_experience = player_years.groupby('Player').agg(years=('penalty_years', 'max')).reset_index()

long_term_players = players_experience[players_experience['years'] >= 10]
long_term_data = penalty_experience[penalty_experience['Player'].isin(long_term_players['Player'])]

longest_experience_player = players_experience.sort_values(by='years', ascending=False).iloc[0]
longest_experience_player
```

```python
example_player = longest_experience_player['Player']
example_player_data = long_term_data[long_term_data['Player'] == example_player]

example_player_performance = example_player_data.groupby('penalty_years').agg(
    penalties_taken=('penalties_taken', 'sum'),
    penalties_scored=('penalties_scored', 'sum'),
    scoring_percentage=('scoring_percentage', 'mean')
).reset_index()

example_player_performance
```

```python
example_player_performance.plot(x='penalty_years', y='scoring_percentage', kind='line')
plt.xlabel('Years of Experience')
plt.ylabel('Average Values')
plt.title('Performance Based on Experience')
plt.show()
```

```python
merged_players['penalty_goals_percentage'] = (merged_players['penalties_scored'] / merged_players['G']) * 100

merged_players[['POS', 'Player', 'penalties_scored', 'G', 'penalty_goals_percentage']]
```

```python
position_penalty_stats = penalty_experience.groupby('POS').agg(
    total_penalties_taken=('penalties_taken', 'sum'),
    total_penalties_scored=('penalties_scored', 'sum')
).reset_index()

position_penalty_stats['penalty_goals_percentage'] = (position_penalty_stats['total_penalties_scored'] / position_penalty_stats['total_penalties_taken']) * 100

position_penalty_stats
```

```python
position_penalty_stats.plot(kind='bar', x='POS', y=['penalty_goals_percentage'])
plt.ylabel('%')
plt.title('Position-wise Performance')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
mypenaltiesgoalkeepers = mypenalties[['Goalkeeper', 'year']].drop_duplicates().reset_index(drop=True)
mypenaltiesgoalkeepers
```

```python
mygoalkeepers = pd.read_csv('all_goalkeepers.csv')
```

```python
mygoalkeepers = mygoalkeepers[mygoalkeepers['Season'] == 'reg']
mygoalkeepers = mygoalkeepers[mygoalkeepers['GP'] > 0]
```

```python
mycolumnslist = ['Year', 'Player', 'GP', 'W%', 'PKG/A', 'SHTS', 'SV', 'Sv%', 'GA', 'GAA']
mygoalkeepers = mygoalkeepers[mycolumnslist]
mygoalkeepers
```

```python
mygoalkeeperstats = pd.merge(mypenaltiesgoalkeepers, mygoalkeepers, left_on=['Goalkeeper', 'year'], right_on=['Player', 'Year'], how='inner')
mygoalkeeperstats
```

```python
mygoalkeeperstats.to_csv('goalkeepers.csv', index=False)
```

```python
goalkeeper_performance = mypenalties.groupby('Goalkeeper').agg(
    penalties_faced=('Outcome', 'size'),
    penalties_saved=('Outcome', lambda x: (x == 'Goalkeeper save').sum())
).reset_index()

goalkeeper_performance['save_probability'] = (goalkeeper_performance['penalties_saved'] / goalkeeper_performance['penalties_faced']) * 100
goalkeeper_performance_overall = goalkeeper_performance.sort_values(by='penalties_faced', ascending=False)

goalkeeper_performance_overall.head(30)
```

```python
penalty_data = mypenalties[['year', 'Goalkeeper', 'Outcome']]

goalkeeper_performance_year = penalty_data.groupby(['year', 'Goalkeeper']).agg(
    penalties_faced=('Outcome', 'size'),
    penalties_saved=('Outcome', lambda x: (x == 'Goalkeeper save').sum())
).reset_index()

goalkeeper_performance_year['save_probability'] = (goalkeeper_performance_year['penalties_saved'] / goalkeeper_performance_year['penalties_faced']) * 100


goalkeeper_performance_year
```

```python
merged_goalkeepers = pd.merge(mygoalkeeperstats, goalkeeper_performance_year, on=['year', 'Goalkeeper'])
merged_goalkeepers
```

```python
yearly_performance = merged_goalkeepers.groupby('year').agg({
    'penalties_faced': 'sum',
    'penalties_saved': 'sum',
    'save_probability': 'mean',
    'SV': 'mean',
    'GAA': 'mean'
}).reset_index()

yearly_performance
```

```python
plt.figure(figsize=(12, 6))
plt.plot(yearly_performance['year'], yearly_performance['save_probability'], label='Penalty Save Percentage', marker='o')
plt.xlabel('Year')
plt.ylabel('Save Percentage')
plt.title('Penalty Save Percentage Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(yearly_performance['year'], yearly_performance['SV'], label='Saves', marker='o')
plt.xlabel('Year')
plt.ylabel('Saves')
plt.title('Average Saves Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(yearly_performance['year'], yearly_performance['GAA'], label='Goals Against Averaged', marker='o')
plt.xlabel('Year')
plt.ylabel('Goals')
plt.title('Goals Scored Against Averaged')
plt.legend()
plt.grid(True)
plt.show()
```

```python
merged_goalkeepers['experience_level'] = pd.cut(merged_goalkeepers['GP'], bins=[0, 10, 20, 30, 40], labels=['0-10', '11-20', '21-30', '31-40'])
experience_performance = merged_goalkeepers.groupby('experience_level').agg(
    penalties_faced=('penalties_faced', 'sum'),
    penalties_saved=('penalties_saved', 'sum'),
    save_probability=('save_probability', 'mean')
).reset_index()

experience_performance
```

```python
plt.figure(figsize=(12, 6))
plt.bar(experience_performance['experience_level'], experience_performance['save_probability'], color='skyblue')
plt.xlabel('Games Played')
plt.ylabel('Average Penalty Save Percentage')
plt.title('Impact of Experience on Penalty Save Performance')
plt.grid(True)
plt.show()
```

```python
experience_performance.plot(kind='bar', x='experience_level', y=['penalties_faced', 'penalties_saved'])
plt.ylabel('Count')
plt.title('Goalkeepers Experience performances on penalties')
plt.show()
```
