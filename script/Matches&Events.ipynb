```python
# Analyzing missing values in 'matches_df'
matches_missing = matches_df.isnull().mean() * 100

# Columns with a high percentage of missing values (threshold set at 60% for this analysis)
high_missing_columns = matches_missing[matches_missing > 60].index

# Removing columns with high percentage of missing values
matches_df_cleaned = matches_df.drop(columns=high_missing_columns)

# Identifying columns with mixed data types by checking for columns where more than one data type is present
mixed_type_columns = matches_df_cleaned.columns[matches_df_cleaned.apply(lambda col: len(col.dropna().map(type).unique()) > 1)]

# Display the columns with high missing values and mixed data types
high_missing_columns, mixed_type_columns
```

```python
events_df = pd.read_csv("events.csv")
```

```python
# Filtering events that contain the word "penalty" in the 'Event' column
penalty_events_df = events_df[events_df['Event'].str.contains('penalty', case=False, na=False)]

# Display the first few rows of the filtered dataframe to understand how penalties are described
penalty_events_df.head()
```

```python
penalty_events_df.value_counts()
```

```python
# Creating a new column to classify if the penalty resulted in a goal
penalty_events_df['Outcome'] = penalty_events_df['Event'].str.contains('Outcome: goal', case=False, na=False)

# Display the modified dataframe to ensure the 'Outcome' column classifies the penalties correctly
penalty_events_df[['Event', 'Outcome']].head()
```

```python
matches_df = pd.read_csv("matches.csv")
```

```python
# Merging the penalty_events_df with matches_df to find match details for each penalty
penalty_match_details_df = penalty_events_df.merge(matches_df, on='id', how='left')

# Display the first few rows of the merged dataframe to review the combined information
penalty_match_details_df[['id', 'Event', 'Outcome', 'home', 'away', 'date', 'venue']].head()
```
